# Introduction

The goal of this analysis is to select and predict CRISPR perturbations of target transcription factor enhancers to see if we can create "cooperative" and "non-cooperative" perturbation modes. We hypothesize by enhancing an important regulatory region with greater cooperative accessibility that we will bolster the ability of this enhancer to act at lower TF concentrations, thus influencing differentiation time/phenotypic readouts of embryonic stem cell lines. 

We will use the differential analysis results from the Xiong et al 2022 time course reinterpretation to select candidate regions and design (based on model predictions of accessibility readout) enhancer inserts of Oct4-Sox2 motifs (and a deletion of Oct4-Sox2 motifs).

```python
def generate_slurm_array(cmds, output_file, simultaneous_jobs = 4):
    total_jobs = len(cmds)
    array_header = ['#!/usr/bin/bash',
                  '#SBATCH --job-name bpnet_training',
                  '#SBATCH --output=slurm_%j.log',
                  f'#SBATCH --array=1-{total_jobs}%{simultaneous_jobs}',
                  '#SBATCH --mem=200gb', 
                  '#SBATCH --time=72:00:00',
                  '#SBATCH --partition=gpu', 
                  '#SBATCH --gres=gpu:a100:1',
                  '#SBATCH --cpus-per-task=10',
                  'source /home/mw2098/.bashrc',
                  'conda deactivate',
                  'conda activate bpreveal_404',
                  f'cd {working_dir}']
    # array_cmds = array_header
    for i,cmd in enumerate(cmds):
        array_cmd = ["if [[ ${{SLURM_ARRAY_TASK_ID}} == {0:d} ]] ; then\n".format(i+1), 
                    "    {0:s}\n".format(cmd),
                    "fi\n\n"]
        array_header = array_header + array_cmd 
        
    with open(output_file, mode='wt') as slurm:
        slurm.write('\n'.join(array_header))
        slurm.write('\n')
```

# Measure predictions with single OS injection

We will now predict several scenarios in different iterations to test whether we can simulate:

1. Loss of function: Mutate Oct4-Sox2 motif of interest to kill accessibility and function within region
2. Gain of function: Mode1: Iteratively add Oct4-Sox2 motifs in a cooperative syntax formation (close-together) to super-additively increase accessibility and function
3. Gain of function: Mode2: Iteratively add Oct4-Sox2 motifs in a non-cooperative syntax formation (spaced apart) to additively increase accessibility and function

We will measure relative accessibility effects over decreasing Oct4 time course models. We hypothesize that:

1. Scenario (1) will now become insensitive to changing levels of Oct4
2. Scenario (2) cooperative arrangements of Oct4-Sox2 will bolster accessibility across decreasing concentrations of Oct4. If so, we hope to see phenotypic differences with proposed CRISPR experiments downstream.
3. Scenario (3) non-cooperative arrangements of Oct4-Sox2 will be subject to similar decreases as the WT predictions over time course, thus showing that cooperativity is essential to maintaining functional in lower TF concentrations.

### Design OS injected sequences

Here, we will inject high-affinity sequences at every position (eliminating +/- 5bp gaps between existing motifs) to see what the predictions will yield. 

```python
inj_range = (-250, 250)

counter = 1
OS_inj_scenarios_df = pd.DataFrame()
for i in range(inj_range[0], inj_range[1], 1):
    
    #Annotate information and injection site
    OS_inj_scenario_df = scenarios_df.copy()
    OS_inj_scenario_df['inj_OS_seq'] = np.where(OS_inj_scenario_df['wt_OS_genomic_strand'] == '-', OS_opt[1], OS_opt[0])
    OS_inj_scenario_df['inj_OS_input_start'] = OS_inj_scenario_df['wt_OS_input_start'] + i  
    OS_inj_scenario_df['inj_OS_input_end'] = OS_inj_scenario_df['wt_OS_input_end'] + i  
    OS_inj_scenario_df['scenario'] = 'OSinj_' + str(counter) + '_' + str(i)

    OS_inj_scenario_df['inj_OS_genomic_start'] = OS_inj_scenario_df['inj_OS_input_start'] + OS_inj_scenario_df['start']  
    OS_inj_scenario_df['inj_OS_genomic_end'] = OS_inj_scenario_df['inj_OS_input_end'] + OS_inj_scenario_df['start'] 

    #Inject motif at new location
    OS_inj_scenario_df['input_seq'] = \
        [s[:OS_inj_scenario_df['inj_OS_input_start'].values[i]] + \
         OS_inj_scenario_df['inj_OS_seq'].values[i] + \
         s[OS_inj_scenario_df['inj_OS_input_end'].values[i]:] \
        for i,s in enumerate(OS_inj_scenario_df['input_seq_wt'].values)]
    
    #Make sure sequences are the same length
    assert len(set([len(OS_inj_scenario_df['input_seq'].values[i]) == len(OS_inj_scenario_df['input_seq_wt'].values[i]) 
                    for i in range(OS_inj_scenario_df.shape[0])]))==1
    OS_inj_scenarios_df = pd.concat([OS_inj_scenarios_df, OS_inj_scenario_df])
```

Filter all injections that overlap with motif locations.

```python
gap_requirement = 5 #how many bp apart should injections be?
motifs_to_check_df = motifs_df[motifs_df.region_id.isin(OS_inj_scenarios_df.region_id.unique())]

#Report original length of predictions
print(OS_inj_scenarios_df.shape)

#Find blacklisted coordinates
blacklisted_coords = []
for i,row in motifs_to_check_df.iterrows():
    for r in range(row.start - gap_requirement, row.end + gap_requirement):
        blacklisted_coords.append(f'{row.chrom}:{r}')

#Find coordinates that injections occupy
inj_coords = []
inj_scenario = []
for i,row in OS_inj_scenarios_df.iterrows():
    for r in range(row.inj_OS_genomic_start - gap_requirement, row.inj_OS_genomic_end + gap_requirement):
        inj_coords.append(f'{row.seqnames}:{r}')
        inj_scenario.append(f'{row.nearest_gene_name}-{row.scenario}')

#Find intersection between the lists
intersections = [i in blacklisted_coords for i in inj_coords]
blacklisted_injections = list(set(np.array(inj_scenario)[intersections]))

#Subset based on these injection overlaps
OS_inj_scenarios_df['scenario_long'] = OS_inj_scenarios_df['nearest_gene_name'] + '-' + OS_inj_scenarios_df['scenario']
OS_inj_scenarios_filt_df = OS_inj_scenarios_df[~OS_inj_scenarios_df.scenario_long.isin(blacklisted_injections)]
print(OS_inj_scenarios_filt_df.shape)
```

## Predict scaling factors across models

Because each model output was trained on separately sequenced experiments, we want to calculate a scaling factor for sequencing depth across models so we can perform a proxy 'RPM-normalization' when comparion outputs. We will plot the actual sequencing depth of the experiments as well to ensure this correlates.

```python
trials = 256
random_seqs = [''.join(list(np.random.choice(['A', 'C', 'G', 'T'], size=input_length, replace=True))) for t in range(trials)]

scaling_factors_df = pd.DataFrame()
for model_name in modeling_design_dict.keys():
    model = load_model(modeling_design_dict[model_name]['model_dir'], 
                           custom_objects = {'multinomialNll' : losses.multinomialNll, 'reweightableMse': losses.dummyMse})
    preds = model.predict(one_hot_encode_sequences(random_seqs))
    tasks = list(modeling_design_dict[model_name]['tasks'])
    for i,task in enumerate(tasks):
        df = pd.DataFrame([np.sum(np.exp(preds[i+len(tasks)]), axis = 0), [task]]).transpose()
        df.columns = ['scaling_factor', 'task']
        df['model_name'] = model_name
        scaling_factors_df = pd.concat([scaling_factors_df, df])
```

```python
scaling_factors_df['model_short'] = scaling_factors_df['model_name'].str.replace('bpnet_osknz', '').str.replace('atac_wt_fold1','').str.replace('_fold1','').str.replace('atac','')
scaling_factors_df['sample_name'] = scaling_factors_df['task'] + scaling_factors_df['model_short']

sequencing_depth_df = pd.read_csv('tsv/seq_depth.tsv',  sep = '\t')
sequencing_depth_df['sample_name'] = sequencing_depth_df['sample_name'].str.replace('GSE174774_', '').str.replace('mesc_','').str.replace('_nexus','')

compare_df = scaling_factors_df.merge(sequencing_depth_df, how = 'left')
compare_df['log_total_reads'] = np.log(compare_df['total_reads'].astype('float64'))
compare_df['log_scaling'] = np.log(compare_df['scaling_factor'].astype('float64'))
```

Confirm that the average random predictions scale with sequencing depth of the joined experiments.

```python
plotnine.options.figure_size = (5,3)

df = compare_df[(compare_df.task=='atac') & (compare_df.sample_name != 'atac')]
seq_plot = (ggplot(data = df, mapping = aes(x = 'log_total_reads', y = 'log_scaling'))+
     geom_text(mapping = aes(label = 'sample_name'))+
 scale_x_continuous(name = 'ln(total sequenced reads)')+
 scale_y_continuous(name = 'ln(total predicted reads)')+
 theme_classic()
)
seq_plot.save(f'{figure_path}/seq_vs_scale.png')
seq_plot.save(f'{figure_path}/seq_vs_scale.pdf')
print(seq_plot)
```

## Predict generated sequences

```python
%%script false --no-raise-error

#Convert logits and logcounts to human-readable ChIP-nexus profile with counts
preds_df = pd.DataFrame()
models_of_interest = ['atac_0h_fold1', 'atac_3h_fold1', 'atac_6h_fold1', 'atac_9h_fold1', 'atac_12h_fold1', 'atac_15h_fold1']


for model_name in tqdm(models_of_interest):
    model = load_model(modeling_design_dict[model_name]['model_dir'], 
                           custom_objects = {'multinomialNll' : losses.multinomialNll, 'reweightableMse': losses.dummyMse})
    preds = model.predict(one_hot_encode_sequences(all_scenarios_df['input_seq'].values))

    tasks = list(modeling_design_dict[model_name]['tasks'])
    for i,task in enumerate(tasks):
        for j,row in all_scenarios_df.iterrows():
            profile = logitsToProfile(logitsAcrossSingleRegion = preds[i][j],
                                      logCountsAcrossSingleRegion = preds[i+len(tasks)][j])

            #Convert to tidy pd.df
            df = pd.DataFrame(profile, columns = [task])
            df['position'] = list(range(df.shape[0]))
            df = df.melt(id_vars = 'position', var_name = 'channel', value_name = 'pred')
            df['genomic_position'] = df['position'] + row.start + flank_length
            df['task'] = task
            df['model_name'] = model_name
            df['scenario_long'] = row.scenario_long

            #Scale results to sequencing depth to do cross-comparisons across related models
            df = df.merge(compare_df[['model_name', 'task', 'scaling_factor']], how = 'left')
            df['pred_norm'] = df['pred']/df['scaling_factor']*1000000
            preds_df = pd.concat([preds_df, df])

summary_df = preds_df.groupby(['task','model_name','scenario_long']).agg({'pred_norm': 'sum'}).reset_index()
summary_df[['nearest_gene_name','region_id','scenario']] = summary_df['scenario_long'].str.split(':', expand=True)
summary_df[['state','inj_n','inj_center_wrt_OS']] = summary_df['scenario'].str.split('_', expand=True)
summary_df['pred_norm'] = summary_df['pred_norm'].astype('float64')

summary_df.to_csv('tsv/genomic/crispr/temp_preds.tsv.gz', sep = '\t')
```

```python
array_cmds = []

for model_name,model_info in modeling_design_dict.items():

    #Set up SHAP parameters for each task.
    shap_template_dict = {
        # 'genome': genome,
        'fasta-file': fasta_path, 
        'input-length': input_length,
        'output-length': output_length,
        'num-shuffles': 20,
        'verbosity': 'DEBUG'
    }

    #For ATAC vs CHIP-nexus, set up SHAP parameters
    if model_info['tasks'][0]=='atac':
        task = 'atac'
        model_suffix = '_residual.model'
        shap_dict = shap_template_dict.copy()
        shap_dict['model-file'] = model_info['model_dir']
        shap_dict['heads'] = 1
        shap_dict['head-id'] = 0
        shap_dict['profile-task-ids'] = [0]
        shap_dict['profile-h5'] = f'shap/crispr/crispr_coop_{model_name}_{task}_profile.h5'
        shap_dict['counts-h5'] = f'shap/crispr/crispr_coop_{model_name}_{task}_counts.h5'
        shap_json = json.dumps(shap_dict, indent=4)
        shap_json_file = f'json/crispr/shapFlat_crispr_coop_{model_name}_{task}.json'
        with open(shap_json_file, 'w') as outfile:
            outfile.write(shap_json)     

        #Set up the SHAP commands
        model_shap = [f'{python_path} {bpreveal_path}/src/interpretFlat.py json/crispr/shapFlat_crispr_coop_{model_name}_{task}.json']
        # cmds = gpu_header + model_shap
        # with open(f'scripts/crispr_coop_bpnet_shap_{model_name}_{task}.slurm', mode='wt') as slurm:
        #     slurm.write('\n'.join(cmds))
        #     slurm.write('\n')
        # print('sbatch ', f'scripts/crispr_coop_bpnet_shap_{model_name}_{task}.slurm')
        #Write to a job array
        cmds_str = '\n'.join(model_shap)
        array_cmd = [cmds_str]
        array_cmds += array_cmd
        
    else:
        model_suffix = '.model'
        shap_dict = shap_template_dict.copy()
        shap_dict['model-file'] = f'models/{model_name}{model_suffix}'
        for head_counter, (task, cov) in enumerate(model_info['cov'].items()):
            shap_dict['heads'] = len(model_info['tasks'])
            shap_dict['head-id'] = head_counter
            shap_dict['profile-task-ids'] = [0,1]
            shap_dict['profile-h5'] = f'shap/crispr/crispr_coop_{model_name}_{task}_profile.h5'
            shap_dict['counts-h5'] = f'shap/crispr/crispr_coop_{model_name}_{task}_counts.h5'
            shap_json = json.dumps(shap_dict, indent=4)
            shap_json_file = f'json/crispr/shapFlat_crispr_coop_{model_name}_{task}.json'
            with open(shap_json_file, 'w') as outfile:
                outfile.write(shap_json)     
    
            #Set up the SHAP commands
            model_shap = [f'{python_path} {bpreveal_path}/src/interpretFlat.py json/crispr/shapFlat_crispr_coop_{model_name}_{task}.json']
            # cmds = gpu_header + model_shap
            # with open(f'scripts/crispr_coop_bpnet_shap_{model_name}_{task}.slurm', mode='wt') as slurm:
            #     slurm.write('\n'.join(cmds))
            #     slurm.write('\n')
            # print('sbatch ', f'scripts/crispr_coop_bpnet_shap_{model_name}_{task}.slurm')   

            #Write to a job array
            cmds_str = '\n'.join(model_shap)
            array_cmd = [cmds_str]
            array_cmds += array_cmd

generate_slurm_array(cmds = array_cmds, output_file = 'scripts/crispr_coop_bpnet_shap.slurm')
print('sbatch scripts/crispr_coop_bpnet_shap.slurm')
```
